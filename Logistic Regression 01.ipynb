```python
df =df[['tenure', 'age', 'address', 'income', 'ed', 'employ', 'equip',   'callcard', 'wireless','churn']]
df.head()
```

```python
df["churn"]=df["churn"].astype("int")
```

```python
## lets Select Independant Variable
X =df.iloc[:,:-1]
X.head()
```

```python
y=df.iloc[:,-1]
y.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=45)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
X_train =np.asanyarray(X_train)
X_train
```

```python
# Used standard scaler for change the data into same range
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train =scaler.fit_transform(X_train)
X_train
```

```python
X_test =scaler.transform(X_test)
```

```python
X_test
```

```python
from sklearn.linear_model import LogisticRegression
clf =LogisticRegression()
clf
```

```python
from sklearn.model_selection import GridSearchCV
Parameter ={"penalty":["l1","l2","elasticnet"],"C":[1,2,3,4,5,10,20,30,50],'max_iter':[100,200,300]}
```

```python
classifier_reg =GridSearchCV(clf,param_grid =Parameter,scoring="accuracy",cv=10)
```

```python
classifier_reg.fit(X_train,y_train)
```

```python
classifier_reg.best_estimator_
```

```python
classifier_reg.best_params_
```

```python
classifier_reg.error_score
```

```python
classifier_reg.best_score_
```

```python
predct =classifier_reg.predict(X_test)
predct
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
print(accuracy_score(predct,y_test))
```

```python
print(classification_report(predct,y_test))
```
